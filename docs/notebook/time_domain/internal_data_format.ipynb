```julia
using SeisPlot, SeisAcoustic

# read velocity and density model
path_vel = joinpath(pwd(), "vel.rsb");
(hdr, vel) = read_RSdata(path_vel);

path_rho = joinpath(pwd(), "rho.rsb");
(hdr, rho) = read_RSdata(path_rho);
```

```julia
SeisPlotTX(vel, wbox=15, hbox=3, cmap="rainbow", vmax=3000, vmin=1000);
SeisPlotTX(rho, wbox=15, hbox=3);
```

```julia
hdr
```

```julia
# vertical and horizontal grid size
dz = hdr.d1; dx = hdr.d2;

# number of PML layers
npml = 20;

# top boundary condition
free_surface = true;


# time step size and maximum modelling length
dt = 0.0015; tmax = 2.25;  # use second as unit

# organize these parameters into a structure
params = TdParams(rho, vel, free_surface, dz, dx, dt, tmax;
         data_format=Float32, fd_flag="taylor", order=6, npml=20, apml=900.);
```

```julia
# initialize a source
src = Source(2, 375, params; ot=0.0, fdom=10.0,
      type_flag="ricker", amp=100000, location_flag="index");


# initialize recordings
irx = collect(1:2:params.nx);
irz = 2 * ones(length(irx));
rec = Recordings(irz, irx, params);
```

```julia
# forward modeling of simultaneous sources
multi_step_forward!(rec, src , params);
SeisPlotTX(rec.p, wbox=10, hbox=10, cmap="seismic");
```

```julia
# save all the pressure field
path_pre = joinpath(pwd(), "pressure.rsb");
multi_step_forward!(path_pre, src, params; save_flag="pressure");
```

```julia
# read the pressure field
(hdr, p) = read_RSdata(path_pre);

# plotting
it = 300; SeisPlotTX(p[:,:,it], wbox=15, hbox=3, cmap="seismic");
it = 600; SeisPlotTX(p[:,:,it], wbox=15, hbox=3, cmap="seismic");
it = 900; SeisPlotTX(p[:,:,it], wbox=15, hbox=3, cmap="seismic");
```
